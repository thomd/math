## Generating correlated random variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
var1 = 10 ** 2
var2 = 8 ** 2
cov = 72
cov_matrix = [[var1, cov],
              [cov, var2]]

data = pd.DataFrame(np.round(np.random.multivariate_normal([170, 85], cov_matrix, 100), decimals=1), columns=['height', 'weight'])
data
```

```python
plt.scatter(data['height'], data['weight']);
```

```python
import seaborn as sns
sns.pairplot(pd.DataFrame(data));
```

```python
np.cov(np.array(data).T)
```

```python
np.corrcoef(np.array(data).T)
```

```python
data.cov()
```

# How to Create a Correlation Matrix using Pandas

```python
import pandas as pd

data = {'A': [45,37,42,35,39],
        'B': [38,31,26,28,33],
        'C': [10,15,17,21,12]
        }

df = pd.DataFrame(data,columns=['A','B','C'])
df
```

```python
import seaborn as sns
sns.pairplot(df);
```

```python
corrMatrix = df.corr()
corrMatrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X_train = np.array([1,2,3,4,5,6]).reshape(-1,1)
y_train = np.array([2,2,4,5,7,7]).reshape(-1,1)

linr = LinearRegression()
linr.fit(X_train, y_train)
print("a={} b={}".format(linr.intercept_, linr.coef_[0]))
```

```python
X_test = np.array([2,4,5,7,9]).reshape(-1,1)
linr.predict(X_test)
```
