Let $P(i)$ be the probability of $i$-th message type, then

$$\text{Entropy} = - \sum_i P(i) \, log_2 P(i)$$

If **entropy is high** (encoding size is big on average), it means we have **many message types with small probabilities**. Hence, every time a new message arrives, youâ€™d expect a different type than previous messages. You may see it as a disorder or uncertainty or unpredictability.

[Source](https://medium.com/activating-robotic-minds/demystifying-entropy-f2c3221e2550)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def uniform_distribution(n):
    return (np.linspace(1, 1, n) / n)

def entropy(P):
    return -1 * np.sum(P * np.log2(P))

x = np.linspace(1, 100, 100)
y = list(map(entropy, list(map(uniform_distribution, x))))

plt.figure(figsize=(12.8, 4.8))
plt.subplot(121)
plt.xlabel('uniform distribution of $n$ elements')
plt.ylabel('entropy')
plt.plot(x, y)
plt.grid()

plt.subplot(122)
plt.xscale('log',basex=2)
plt.plot(x, y)
plt.grid()
```
